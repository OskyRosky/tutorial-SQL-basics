## Creación de base de datos + tablas 

```python
# Establece la conexión con la base de datos
conexion = mysql.connector.connect(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="amoaDios1986!"
)

# Crea un cursor para interactuar con la base de datos
cursor = conexion.cursor()
```

```python
# Crea la base de datos "Mundo dato"
cursor.execute("CREATE DATABASE IF NOT EXISTS `Mundo Data`")

# Selecciona la base de datos "Mundo dato"
cursor.execute("USE `Mundo Data`")
```

```python
# Crea la tabla Formación
cursor.execute("""
CREATE TABLE IF NOT EXISTS Formación (
    Id INT AUTO_INCREMENT PRIMARY KEY,
    tema VARCHAR(255) NOT NULL,
    horas INT NOT NULL,
    nivel VARCHAR(50) NOT NULL,
    costo DECIMAL(10,2) NOT NULL
)
""")

# Crea la tabla Clientes
cursor.execute("""
CREATE TABLE IF NOT EXISTS Clientes (
    Id INT AUTO_INCREMENT PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL,
    Ubicación VARCHAR(255) NOT NULL,
    Pago DECIMAL(10,2) NOT NULL,
    Fecha DATE NOT NULL
)
""")

# Crea la tabla Personal
cursor.execute("""
CREATE TABLE IF NOT EXISTS Personal (
    Id INT AUTO_INCREMENT PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL,
    Apellidos VARCHAR(255) NOT NULL,
    Edad INT NOT NULL,
    Sexo CHAR(1) NOT NULL,
    Fecha_de_nacimiento DATE NOT NULL,
    Salario DECIMAL(10,2) NOT NULL
)
""")
```

```python
cursor.close()
conexion.close()

# Este código crea la base de datos "Mundo dato" y las tablas "Formación", "Clientes" y "Personal" con las columnas especificadas. 
# Si alguna de las tablas ya existe, el código no la volverá a crear gracias a la cláusula IF NOT EXISTS.
```

## Insertar  datos a la tabla de una base de datos

```python
# Función para obtener una fecha aleatoria de los últimos 5 años
def random_date():
    start_date = datetime.now() - timedelta(days=5*365)
    random_days = random.randint(0, 5*365)
    return start_date + timedelta(days=random_days)
```

```python
# Establece la conexión con la base de datos
conexion = mysql.connector.connect(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="amoaDios1986!",
    database="Mundo Data"
)
cursor = conexion.cursor()
```

```python
# Inserta datos en la tabla Formación
for _ in range(10):
    tema = random.choice(['Programación', 'Diseño Gráfico', 'Marketing', 'Base de Datos', 'Redes'])
    horas = random.randint(5, 60)
    nivel = random.choice(['Básico', 'Intermedio', 'Avanzado'])
    costo = round(random.uniform(50, 300), 2)
    cursor.execute("INSERT INTO Formación (tema, horas, nivel, costo) VALUES (%s, %s, %s, %s)", (tema, horas, nivel, costo))
```

```python
# Inserta datos en la tabla Clientes
nombres = ['Juan', 'María', 'Carlos', 'Lucía', 'Diego', 'Sofía', 'Miguel', 'Ana', 'Pedro', 'Isabel']
ubicaciones = ['Bogotá', 'Lima', 'Santiago', 'Buenos Aires', 'Ciudad de México']
for nombre in nombres:
    ubicacion = random.choice(ubicaciones)
    pago = round(random.uniform(50, 300), 2)
    fecha = random_date()
    cursor.execute("INSERT INTO Clientes (Nombre, Ubicación, Pago, Fecha) VALUES (%s, %s, %s, %s)", (nombre, ubicacion, pago, fecha))
```

```python
# Inserta datos en la tabla Personal
apellidos = ['Pérez', 'García', 'Rodríguez', 'Fernández', 'Martínez', 'Guzmán', 'López', 'Torres', 'Gómez', 'Castro']
sexos = ['M', 'F']
for nombre, apellido in zip(nombres, apellidos):
    edad = random.randint(20, 60)
    sexo = random.choice(sexos)
    fecha_nacimiento = random_date()
    salario = round(random.uniform(800, 1500), 2)
    cursor.execute("INSERT INTO Personal (Nombre, Apellidos, Edad, Sexo, Fecha_de_nacimiento, Salario) VALUES (%s, %s, %s, %s, %s, %s)", 
                   (nombre, apellido, edad, sexo, fecha_nacimiento, salario))
```

```python
# Confirma los cambios y cierra la conexión
conexion.commit()
cursor.close()
conexion.close()
```
